```python
restaurent_ratings_dataset['Cuisines'] = restaurent_ratings_dataset['Cuisines'].fillna('unknown')
restaurent_ratings_dataset['Cuisines']
```

```python
restaurent_ratings_dataset.isnull().sum()
```

```python

restaurent_ratings_dataset = restaurent_ratings_dataset.drop_duplicates()
restaurent_ratings_dataset
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

text_features = ["Restaurant Name", "City", "Address", "Locality", "Locality Verbose", 
                 "Cuisines", "Currency", "Rating color", "Rating text", "Votes"]

for col in text_features:
    restaurent_ratings_dataset[col] = le.fit_transform(restaurent_ratings_dataset[col].astype(str))  # Convert to string if needed
restaurent_ratings_dataset
```

```python
binary_columns = ["Has Table booking", "Has Online delivery", "Is delivering now", "Switch to order menu"]

for col in binary_columns:
    restaurent_ratings_dataset[col] = restaurent_ratings_dataset[col].map({"Yes": 1, "No": 0})
restaurent_ratings_dataset
```

```python
x=restaurent_ratings_dataset.drop(columns=["Aggregate rating"])
x
```

```python
y=restaurent_ratings_dataset["Aggregate rating"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print("MSE:",mse)
print("R2:",r2)
```
